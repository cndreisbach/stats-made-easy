```python
import random
```

```python
def heads(n=100):
    total = 0
    for _ in range(n):
        if random.random() > 0.5:
            total += 1
    return total
```

```python
heads()
```

```python
results = [heads() for _ in range(10000)]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.distplot(results, bins=20)
```
